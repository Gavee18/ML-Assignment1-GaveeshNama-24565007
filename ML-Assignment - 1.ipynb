```python
# Univarient Analysis

# Country vice number of transactions

df['Country'].value_counts().plot.bar()
```

```python
# Number of different type of transactions

df['Transaction Type'].value_counts().plot.bar()
```

```python
# Which industry have most transactions

df['Industry'].value_counts().plot.bar()
```

```python
# Pie chart of number of Transactions reported to authorities

df['Reported by Authority'].value_counts().plot.pie()
```

```python
# Month vice distribution of number of Transactions

df['Date of Transaction'] = pd.to_datetime(df['Date of Transaction'])
df['month'] = df['Date of Transaction'].dt.month_name()


df['month'].value_counts().plot.bar()
```

```python
# Bivarient analysis

import matplotlib.pyplot as plt
import seaborn as sns

# How much money transfered in which month (Numerical - Catagorical)

plt.figure(figsize=(20,6))
sns.barplot(x=df['month'], y=df['Amount (USD)'])
plt.title('Transaction Amount vs. Date')
plt.show()
```

```python
# how much money in which Industry (Numerical - Catagorical)

plt.figure(figsize=(12,6))
sns.barplot(x=df['Industry'], y=df['Amount (USD)'])
plt.title('Scatter Plot: Amount (USD) vs. Industry')
plt.show()
```

```python
# Transaction amount accross countries (Numerical - Catagorical)

plt.figure(figsize=(15,6))
sns.boxplot(x='Country', y='Amount (USD)', data=df)
plt.xticks(rotation=90)
plt.title('Transaction Amount across Countries')
plt.show()
```

```python
# Heatmap of cases reported by authorities and sources of money (Catagorical - Catagorical)

heatmap_data = pd.crosstab(df['Reported by Authority'], df['Source of Money'])
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Cases Reported by Authority and Source of Money')
plt.xlabel('Source of Money')
plt.ylabel('Reported by Authority')
plt.show()
```
